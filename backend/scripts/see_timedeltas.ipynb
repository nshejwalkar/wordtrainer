```python
import matplotlib.pyplot as plt

# Load PTS times from ffprobe output
with open("../csvs/pts.csv") as f:
   pts = [float(line.strip()) for line in f if line.strip()]

# Calculate time deltas between successive frames
deltas = [pts[i+1] - pts[i] for i in range(len(pts)-1)]

# Plot
plt.figure(figsize=(12, 4))
plt.plot(deltas, marker='.', linestyle='-', label='ΔTime between frames')
plt.axhline(y=1/60, color='gray', linestyle='--', label='Ideal 60 FPS (16.67ms)')
plt.xlabel('Frame Index')
plt.ylabel('Frame Time Δ (seconds)')
plt.title('Frame Spacing (VFR Check)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
